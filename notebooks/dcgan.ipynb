```python
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt

imgs = np.load('../log/pca_feature_PriorTarget/dcgan_new2_0.01/images/1000.npy')
imgs.shape
```

```python
embeds = np.load(Path('../data/cifar10/pca.npz').open('rb'))
mean = embeds['mean']
components = embeds['components']
cov_eigs = embeds['cov_eigs']
```

```python
plt.imshow((components[90] * 0.5 + 0.5).reshape(3, 32, 32).transpose(1, 2, 0))
```

```python
x = ((imgs.reshape(len(imgs), -1) - 0.5) / 0.5 - mean[None, :]) @ components.T / cov_eigs
```
